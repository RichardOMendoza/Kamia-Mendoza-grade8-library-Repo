```python
import requests
import json
url = "https://raw.githubusercontent.com/RichardOMendoza/Kamia-Mendoza-grade8-library-Repo/refs/heads/main/books.json"
response = requests.get(url)
books = response.json()
print(books)
```

```python
import firebase_admin
from firebase_admin import credentials, db
# Load the private key
cred = credentials.Certificate("/content/firebase_key.json")
# Initialize the app with your database URL
firebase_admin.initialize_app(cred, {
 "databaseURL": "https://kamiamendoza-grade8-library-db-default-rtdb.asia-southeast1.firebasedatabase.app/"
})
print("Firebase connected successfully!")
```

```python
import requests
import json
"""
The 'students' variable is already loaded from the URL in a previous cell.
Therefore, we can directly use the 'students' variable.
"""
data = books
print("JSON file loaded")
ref = db.reference("books")
for books in data["books"]:
  ref.child(str(books["id"])).set(books)
print("Data uploaded successfully!")
```

```python
books_ref = db.reference("books")
books_data = books_ref.get()
print(books_data)
```

```python
for books_id, info in books_data.items():
 print(info["title"])
```

```python
for books_id, info in books_data.items():
 print(info["title"], " Availability:" ,info['details']["available"])
```

```python
total = 0

for books_id, info in books_data.items():
  total += info['details']["copies"]
print("The number of books in the library are :", total)
```
